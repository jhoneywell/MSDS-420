```python
sales_over_thousand=custtrans.query('YTD_SALES_2009>1000')
custtrans.groupby('YTD_SALES_2009').size()
print(sales_over_thousand)
```

```python
sales_over_thousand_rt=sales_over_thousand.query('CHANNEL_ACQUISITION=="RT"')
print(sales_over_thousand_rt)
```

```python
sales_over_thousand_rt_zip=sales_over_thousand_rt.query('ZIP==60056').ZIP.count()
print("There are {} records with sales > 1000, channel acquisition equal to RT, and zip code 60056.".format(sales_over_thousand_rt_zip))
```
